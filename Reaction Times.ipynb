# Segmented Psychophysics

## In this notebook i will describe the reaction times by groups. 
## Groups will be done according to the hits, misses, correct rejections and stimulus misunderstood 

```python
from numpy import loadtxt, ones, arange, mean, std, exp, sqrt, pi
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 12, 'font.weight':'bold', 'font.style':'normal'})
data=loadtxt('/run/media/sparra/AENHA/BaseDatosKarlitosNatsushiRR032/RS/Psicofisica_seccionada.csv')
```

The first question to to answer is if there is difference among reaction times in hits depending on the type of stimulus.

An appropiate answer will be gived to us by scatter plots related to every one of the stimulus.

```python
RX_hits_T=data[:, 5]
RX_hits_A=data[:, 11]
RX_hits_NS=data[:, 26]
RX_miss_T=data[:, 36]
RX_miss_A=data[:, 41]
RX_missT_NS=data[:, 15] #Falsas alarmas táctiles
RX_missA_NS=data[:, 20] #Falsas alarmas táctiles
RX_confT_A=data[:, 29] #Missconception from tactile to auditive
RX_confA_T=data[:, 32] #Missconception from auditive to tactile
plt.figure(0, figsize=(15, 16))
ax1=plt.subplot(2, 1, 1)
ax1.scatter(ones(len(RX_hits_A)), RX_hits_A, c='red', label='Auditive H')
ax1.scatter(ones(len(RX_hits_NS))*2, RX_hits_NS, c='black', label='Not-stimulus H')
ax1.scatter(ones(len(RX_hits_T))*3, RX_hits_T, c='blue', label='Tactil H')
ax1.legend(loc='best', frameon=False)
ax1.set_title('Reaction times', fontsize=16, fontweight='bold')
ax1.set_ylabel('Reaction time [s]', fontsize=14, fontweight='bold')
#ax1.set_xlabel('Stimuli', fontsize=14, fontweight='bold')
ax1.set_xticks(arange(1, 4), ('Auditive', 'Non-stimulus', 'Tactile'))
#
ax1.scatter(ones(len(RX_miss_A))*4, RX_miss_A, c='orange', label='miss A')
ax1.scatter(ones(len(RX_missT_NS))*5, RX_missT_NS, c='cyan', label='Tactile FA')
ax1.scatter(ones(len(RX_missA_NS))*6, RX_missA_NS, c='pink', label='Auditive FA')
ax1.scatter(ones(len(RX_miss_T))*7, RX_miss_T, c='magenta', label='miss T')
#
ax1.scatter(ones(len(RX_confT_A))*8, RX_confT_A, c='gray', label='MCT-A')
ax1.scatter(ones(len(RX_confA_T))*9, RX_confA_T, c='brown', label='MCA-T')
ax1.legend(loc='best', frameon=False)

```

The uncertainty set was always the second set to be presented during experiments, then reaction times could reflect the animals's confidence. It is remarkable that reaction times does not reflect a specific button. If that was the case, answers related to the same button should get similar values, and such things does not happens.
