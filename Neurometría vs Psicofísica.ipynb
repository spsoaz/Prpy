# Neurometría vs Psicofísica


En estas notas se realizarán análisis de la neurometría contra la psicofísica mediante el uso de la sensibilidad d prima d'

```python
import numpy as np
import matplotlib.pyplot as plt

sA=np.loadtxt('/run/media/sparra/AENHA/BaseDatosKarlitosNatsushiRR032/Prpy/sensitivity_A1_RR032.csv', dtype=float)
sT=np.loadtxt('/run/media/sparra/AENHA/BaseDatosKarlitosNatsushiRR032/Prpy/sensitivity_S1_RR032.csv', dtype=float)
dA=np.divide(sA[:, 5]-sA[:,4], sA[:, 4])
dT=np.divide(sT[:, 5]-sT[:,4], sT[:, 4])
plt.figure(0, figsize=(10, 10))
plt.plot(sA[:, 4], sA[:, 6], 'k*' )
plt.plot(sT[:, 4], sT[:, 6], 'k*' )
plt.xlabel('Sensitivity for Tactile stimuli d´', fontsize=12, fontweight='bold')
plt.ylabel('Sensitivity for Acoustic stimuli d´', fontsize=12, fontweight='bold')
x=np.arange(1.1, 3.5, 0.01)
plt.plot(x, x, c='red', label='equal sensitivity')
plt.legend(loc='best', fontsize=8)
plt.grid(True)
#Proportion of rewards
plt.figure(1, figsize=(10, 10))
plt.plot(sA[:, 8], sA[:, 10], 'k*' )
plt.plot(sT[:, 8], sT[:, 10], 'k*' )
plt.xlabel('Proportion of Hits for Tactile flutter', fontsize=12, fontweight='bold')
plt.ylabel('Proportion of Hits for Acoustic flutter', fontsize=12, fontweight='bold')
x=np.arange(0, 1, 0.01)
plt.plot(x, x, c='red', label='equality')
plt.legend(loc='best', fontsize=8)
plt.grid(True)
```

```python
plt.figure(2, figsize=(10, 10))
plt.hist(dT, label='Tactile flutter')
plt.hist(dA, label='Acoustic flutter')
plt.xlabel('Neurometric sensitivity error', fontsize=12, fontweight='bold')
plt.legend(loc='best', fontsize=8)
```
